El siguiente código crea un mapa interactivo que responde a la pregunta:

**¿Cómo varían los precios medios de los alojamientos de Airbnb en Barcelona según el barrio, y cuál es la concentración de alojamientos en cada barrio?**

### Detalles del Código:

- **Agrupación de Datos**:
  - Agrupa los datos por barrio para calcular la media de precio (`avg_price`) y el número de alojamientos (`num_listings`) en cada barrio.
  - Calcula la media de latitud y longitud para representar la ubicación central de cada barrio.

- **Visualización Interactiva**:
  - Usa `scatter_mapbox` de Plotly Express para crear un mapa de dispersión.
  - **Color**: Representa la media de precio en cada barrio, lo que permite visualizar cómo varían los precios en diferentes partes de la ciudad.
  - **Tamaño**: Representa el número de alojamientos en cada barrio, mostrando la concentración de alojamientos de manera proporcional (cuanto más grande el punto, más alojamientos hay en ese barrio).

### Justificación de la Elección del Gráfico:
- **Expresividad y Efectividad**: Un mapa de dispersión interactivo es eficaz para mostrar tanto la variabilidad de los precios como la densidad de los alojamientos en diferentes barrios. Utilizar el color para representar los precios y el tamaño para la cantidad de alojamientos facilita una comprensión visual rápida y clara de estos aspectos.
- **Interactividad**: La posibilidad de interactuar con el mapa (zoom, desplazamiento, información detallada en tooltips) enriquece la exploración de los datos, permitiendo identificar fácilmente barrios con precios más altos o bajos y áreas con mayor o menor concentración de alojamientos.

### Uso en el Proyecto:
Este gráfico sería una parte esencial de tu visualización final, proporcionando una vista global y detallada de cómo los precios de los alojamientos y su concentración varían en Barcelona, ayudando a responder preguntas sobre patrones geográficos y diferencias entre barrios en términos de precio y disponibilidad de alojamientos.

```python
import pandas as pd


# Agrupar por barrio y calcular la media de precio y el número de alojamientos
grouped_data = data.groupby('neighbourhood').agg({
    'price': 'mean',
    'id': 'count',
    'latitude': 'mean',
    'longitude': 'mean'
}).reset_index()

# Renombrar columnas para mayor claridad
grouped_data.rename(columns={'id': 'num_listings', 'price': 'avg_price'}, inplace=True)

# Mostrar las primeras filas del DataFrame agrupado
print(grouped_data.head())
```

```python
import plotly.express as px

# Crear el mapa de dispersión interactivo
fig = px.scatter_mapbox(grouped_data,
                        lat='latitude',
                        lon='longitude',
                        hover_name='neighbourhood',
                        hover_data=['avg_price', 'num_listings'],
                        color='avg_price',
                        color_continuous_scale='emrld',
                        size='num_listings',
                        size_max=60,
                        zoom=11,
                        height=600,
                        mapbox_style='carto-positron',
                        title='Media de Precio de Alojamientos por Barrio en Barcelona')

# Mostrar el mapa
fig.show()
```

El código proporcionado responde a la pregunta:

**¿Cuál es el precio medio de los alojamientos por barrio en Barcelona?**

### Detalles del Código:

- **Agrupación y Cálculo**:
  - Agrupa los datos por 'neighbourhood' y calcula el precio medio (`avg_price`) para cada barrio.
  
- **Creación del Gráfico**:
  - Utiliza `plotly.express.bar` para crear un gráfico de barras interactivas.
  - El color de las barras se asigna según el precio medio, utilizando la escala de colores 'amp'.
  - El tamaño y formato del tooltip se personalizan para mostrar el precio medio con dos decimales y el símbolo de euro (€).

### Características del Gráfico:

- **Tooltip Personalizado**:
  - El tooltip muestra el precio medio de cada barrio con el formato `Precio Medio = X.XX €`.
  
- **Interactividad**:
  - Al pasar el cursor sobre una barra, se despliega el precio medio con formato específico, mejorando la comprensión y la visualización de los datos.

### Justificación de la Elección del Gráfico:

- **Claridad**:
  - La visualización con barras permite comparar fácilmente los precios medios entre los diferentes barrios.
- **Interactividad**:
  - La capacidad de interactuar con el gráfico y ver los precios específicos al pasar el cursor mejora la experiencia del usuario y facilita la exploración de los datos.

### Uso en el Proyecto:

Este gráfico es útil para entender rápidamente cómo se distribuyen los precios medios de los alojamientos en Barcelona, destacando las diferencias entre barrios. Esto es especialmente valioso para decisiones de inversión, selección de alojamiento o estudios de mercado.

```python
import pandas as pd
import plotly.express as px

# Cargar los datos
# data = pd.read_csv('Cleaned_airbnb_barcelona_no_outliers.csv')

# Agrupar por barrio y calcular el precio medio
avg_price_by_neighbourhood = data.groupby('neighbourhood')['price'].mean().reset_index()

# Ordenar los datos por precio medio (opcional)
avg_price_by_neighbourhood = avg_price_by_neighbourhood.sort_values(by='price', ascending=False)

# Crear el gráfico de barras interactivas con formato de tooltip personalizado
fig = px.bar(avg_price_by_neighbourhood,
             x='neighbourhood',
             y='price',
             color='price',
             color_continuous_scale='magenta',
             hover_name='neighbourhood',
             title='Precio Medio por Barrio en Barcelona',
             labels={'neighbourhood': 'Barrio', 'price': 'Precio Medio'})

# Personalizar el formato del tooltip
fig.update_traces(hovertemplate='<b>%{y:.2f} €</b>')

# Ajustar el diseño del gráfico
fig.update_layout(xaxis_title='Barrio', yaxis_title='Precio Medio',
                  xaxis_tickangle=-45, height=600)

# Mostrar el gráfico
fig.show()
```

El código proporcionado responde a la pregunta:

**¿Cómo varía la disponibilidad promedio de alojamientos y el precio medio por barrio en Barcelona a lo largo de diferentes períodos?**

### Detalles del Código:

- **Preparación de los Datos**:
  - Se cargan los datos del archivo CSV que contiene información sobre alojamientos en Barcelona, incluyendo las columnas de disponibilidad para diferentes períodos (`availability_30`, `availability_60`, `availability_90`, `availability_365`) y el precio (`price`).
  - Se agrupan los datos por barrio para calcular la disponibilidad promedio y el precio medio.

- **Creación del Gráfico Interactivo**:
  - Utiliza Plotly Express para crear un gráfico de líneas interactivo.
  - Las líneas representan la disponibilidad promedio de alojamientos en cada barrio a lo largo de 30, 60, 90 y 365 días.
  - Se añade una línea adicional que representa el precio medio de los alojamientos por barrio.
  - Se configura un tooltip personalizado para mostrar el precio medio con el símbolo de euro (€) y dos decimales al pasar el cursor sobre la línea correspondiente al precio medio.

### Justificación del Gráfico de Líneas Interactivo:

- **Comparación de Tendencias**: Permite comparar visualmente cómo cambia la disponibilidad de alojamientos y el precio medio en diferentes barrios de Barcelona a lo largo de períodos diferentes.
- **Interactividad**: Facilita la exploración detallada de datos al permitir al usuario ver valores específicos al pasar el cursor sobre las líneas.
- **Información Completa**: Combina dos aspectos clave (disponibilidad y precio) en una sola visualización, proporcionando una imagen completa de la situación del mercado de alojamientos en cada barrio.

### Aplicación en el Proyecto:

Este tipo de visualización es valioso para analizar las dinámicas del mercado de alojamientos en Barcelona, ayudando a identificar patrones estacionales en la disponibilidad y comprender cómo estos pueden afectar o estar relacionados con los precios medios en diferentes áreas de la ciudad. Es útil para tomar decisiones estratégicas en sectores como el turismo, la inversión inmobiliaria o la planificación urbana.

```python
import pandas as pd
import plotly.express as px

# Cargar los datos
# data = pd.read_csv('Cleaned_airbnb_barcelona_no_outliers.csv')

# Seleccionar columnas relevantes para disponibilidad y precio
columns = ['neighbourhood', 'availability_30', 'availability_60', 'availability_90', 'availability_365', 'price']
combined_data = data[columns].copy()

# Calcular disponibilidad promedio por barrio
availability_avg_by_neighbourhood = combined_data.groupby('neighbourhood').mean().reset_index()

# Calcular precio medio por barrio
price_avg_by_neighbourhood = combined_data.groupby('neighbourhood')['price'].mean().reset_index()

# Mostrar las primeras filas para verificar
print(availability_avg_by_neighbourhood.head())
print(price_avg_by_neighbourhood.head())
```

```python
# Configurar el gráfico de línea interactivo con disponibilidad y precio medio
fig = px.line(availability_avg_by_neighbourhood,
              x='neighbourhood',
              y=['availability_30', 'availability_60', 'availability_90', 'availability_365'],
              title='Disponibilidad y Precio Medio de Alojamientos por Barrio en Barcelona',
              labels={'neighbourhood': 'Barrio', 'value': 'Disponibilidad/Precio Medio', 'variable': 'Días'},
              hover_name='neighbourhood')

# Añadir la línea para el precio medio
fig.add_scatter(x=price_avg_by_neighbourhood['neighbourhood'],
                y=price_avg_by_neighbourhood['price'],
                mode='lines',
                name='Precio Medio',
                hovertemplate='<b>%{y:.2f} €</b>',
                line=dict(color='black', dash='dot'))

# Ajustar el diseño del gráfico
fig.update_layout(xaxis_title='Barrio', yaxis_title='Disponibilidad/Precio Medio',
                  xaxis_tickangle=-45, height=600)

# Mostrar el gráfico
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Cargar los datos
# data = pd.read_csv('Cleaned_airbnb_barcelona_no_outliers.csv')

# Calcular la frecuencia del tipo de alojamiento
property_type_counts = data['property_type'].value_counts().reset_index()
property_type_counts.columns = ['property_type', 'count']

# Calcular el precio medio por tipo de alojamiento
average_price_by_type = data.groupby('property_type')['price'].mean().reset_index()
average_price_by_type.columns = ['property_type', 'average_price']

# Combinar los datos en un solo DataFrame
combined_data = pd.merge(property_type_counts, average_price_by_type, on='property_type')

# Ordenar por frecuencia (count) de más a menos común
combined_data = combined_data.sort_values(by='count', ascending=False)

# Crear el treemap interactivo
fig = px.treemap(combined_data,
                 path=['property_type'],
                 values='count',
                 color='average_price',
                 color_continuous_scale='amp',
                 title='Frecuencia del Tipo de Alojamiento y Precio Medio en Barcelona',
                 hover_data={'count': True, 'average_price': ':.2f €'})

# Ajustar el diseño del treemap
fig.update_layout(height=800)

# Mostrar el treemap
fig.show()
```
